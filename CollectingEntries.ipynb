```python
#importing libraries
import requests
import csv
from bs4 import BeautifulSoup as bs

#site we will collect entries from is eksisozluk
site= 'https://eksisozluk.com/'
#topic that we are interested in
gundemTopics=['vladimir putin','uzay çetin']
allEntries=[]
```

```python
#method to collect 1500 entries from the topic in eksisozluk
def vericek(baslik):
    #header to avoid from botprotector
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    #creating url to use with beautiful soup. url is created according to eksisozluk's url draft.
    r = requests.get(site + baslik, headers=headers)
    url=r.url
    page='?p='
    pageNum=1
    url=url+page+str(pageNum)
    
    #if page doesn't exists 
    if r.status_code != 200:
        print ('bulamadim')
    #if page exists:
    else:
        while(r.status_code==200):
            #increasing pagenum in every step to get next page's entries
            url=url[:url.index("=")+1]+str(pageNum)
            r = requests.get(url,headers=headers)
            #parsing html to get only entries from the page.
            #the html tag that used is 'entry-item-list' 
            soup = bs(r.content, 'html.parser')
            if soup.find(id= 'entry-item-list'):
                entryler =soup.find(id= 'entry-item-list').find_all('li')
                #since entry-item-list is a list of 10 entries in each page; seperate them and collect in allEntries.
                for entry in entryler:
                    icerik = entry.find(class_ = 'content').get_text(strip=True)
                    print ('{}'.format( icerik))
                    allEntries.append(icerik)
                #should stop when we get 1500 entries. in page 150 there will be 1500 entries collected already.
                if(pageNum==150):
                    break
                pageNum=pageNum+1
```

```python
#showing all entries that are collected from vladimir putin &uzay çetin topics in eksisozluk.com
for topic in gundemTopics:
    vericek(topic)
```

```python
#number of entries collected
len(allEntries)
```

```python
#creating csv file named 'entries.csv' and writing all these entries in this csv file
with open('entries.csv', 'w', newline='') as writeFile:
    writer = csv.writer(writeFile, delimiter='\n',quotechar='"', quoting=csv.QUOTE_ALL)
    writer.writerow(allEntries)
```
